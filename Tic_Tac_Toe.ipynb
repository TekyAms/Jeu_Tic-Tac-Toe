```python
from random import randrange


def display_board(board):
	print("+-------" * 3,"+", sep="")
	for row in range(3):
		print("|       " * 3,"|", sep="")
		for col in range(3):
			print("|   " + str(board[row][col]) + "   ", end="")
		print("|")
		print("|       " * 3,"|",sep="")
		print("+-------" * 3,"+",sep="")


def entrer_coup(board):
	ok = False	# fausse hypothèse – nous en avons besoin pour entrer dans la boucle
	while not ok:
		coup = input("Entrez votre coup: ")
		ok = len(coup) == 1 and coup >= '1' and coup <= '9' # la saisie de l'utilisateur est-elle valide ?
		if not ok:
			print("Mauvais coup - refaites la saisie!") # non, elle ne l'est pas - refaites la saisie
			continue
		coup = int(coup) - 1 	# numéro de cellule de 0 à 8
		row = coup // 3 	# ligne de cellule
		col = coup % 3		# column de cellule
		sign = board[row][col]	#  vérifiez la case sélectionnée
		ok = sign not in ['O','X']
		if not ok:	# c'est occupé - revenez à la saisie
			print("Champ déjà occupé - refaites la saisie!")
			continue
	board[row][col] = 'O' 	# mettez « 0 » sur la case sélectionnée


def make_list_of_free_fields(board):
	free = []
	for row in range(3): # itérer sur les lignes
		for col in range(3): # itérer sur les colonnes
			if board[row][col] not in ['O','X']: # la cellule est-elle libre ?
				free.append((row,col)) # oui, elle l'est - ajoutez un nouveau tuple à la liste
	return free


def victory_for(board,sgn):
	if sgn == "X":	# cherchons-nous X ?
		who = 'me'	# oui - c'est du côté de l'ordinateur
	elif sgn == "O": # ... ou pour O?
		who = 'you'	# oui - c'est notre côté
	else:
		who = None	# nous ne devrions pas tomber ici !
	cross1 = cross2 = True  # pour les diagonales
	for rc in range(3):
		if board[rc][0] == sgn and board[rc][1] == sgn and board[rc][2] == sgn:	# vérifiez la ligne rc
			return who
		if board[0][rc] == sgn and board[1][rc] == sgn and board[2][rc] == sgn: # vérifiez la column rc
			return who
		if board[rc][rc] != sgn: # vérifiez 1st diagonal
			cross1 = False
		if board[2 - rc][2 - rc] != sgn: # vérifiez 2nd diagonal
			cross2 = False
	if cross1 or cross2:
		return who
	return None


def draw_coup(board):
	free = make_list_of_free_fields(board) # faites une liste de champs libres
	cnt = len(free)
	if cnt > 0:
		this = randrange(cnt)
		row, col = free[this]
		board[row][col] = 'X'


board = [ [3 * j + i + 1 for i in range(3)] for j in range(3) ]
board[1][1] = 'X' # mettez le premier 'X' au milieu
free = make_list_of_free_fields(board)
human_turn = True # à quel tour est-ce maintenant ?
while len(free):
	display_board(board)
	if human_turn:
		entrer_coup(board)
		victor = victory_for(board,'O')
	else:
		draw_coup(board)
		victor = victory_for(board,'X')
	if victor != None:
		break
	human_turn = not human_turn
	free = make_list_of_free_fields(board)

display_board(board)
if victor == 'you':
	print("You won!")
elif victor == 'me':
	print("I won")
else:
	print("Tie!")
```
